## Visuospatial 1, Trail Making

Patients taking the test must be able to (without additional prompt) continue the given 1 -> A -> 2 -> B... pattern.

If the patient can draw a continuous line 1A2B3C4D5E, 1 point should be added to their MoCA score.

## Visuospatial 2, Copy lineart Cube / Bed

Patients must be able to replicate a lineart cube to sufficient recognisability.

(for the purposes of MedHack, I don't think the specifics of how recognisable are particularly important. I think the ability to distinguish with some degree of accuracy between clear cubes and clear random lines is sufficient for a proof of concept)

If patient is able to draw lineart cube, 1 point should be added to their MoCA score

### Potential method:
CNN trained on dataset of Random Lines vs Lineart Cubes.

## Visuospatial 3, Draw a Clock

Patients must be able to draw a clock which generally resembles a clock, has numbers represented at approximately correct angles from centre, and hands showing the specified time (in the MoCA test found on the google drive, this is 5 part 10)

For each of the following, 1 point should be added to the patient's MoCA score. 
- The Clock contains an approximately circular face
- The numbers are placed at approximately equidistant angles and appropriate positions
- The hands are placed at approximately correct angles for the specified time.

### Potential method

Online, a pre-existing dataset of clocks (~47,000) is available.
